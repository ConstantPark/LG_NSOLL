```python
import pandas as pd
from scipy.io import arff

data = arff.loadarff('../input/lge-project/FordA_TRAIN.arff')
df = pd.DataFrame(data[0])
df.head()
```

```python
import pandas as pd

for col in df:
    if pd.api.types.is_object_dtype(df[col]):
        try:
            df[col] = df[col].astype(int)
        except ValueError:
            pass
```
